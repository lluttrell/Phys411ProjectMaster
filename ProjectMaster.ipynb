## Interpolate over NaN's and save as text files

### minute data

```python
#import data
dc = np.genfromtxt('DeepCoveNew.txt')
jb = np.genfromtxt('JamesBayNew.txt')
jm = np.genfromtxt('JohnMuirNew.txt')
uv = np.genfromtxt('UVicISCnew.txt')

#save original time points
tdc = dc[:,0]
tjb = jb[:,0]
tjm = jm[:,0]
tuv = uv[:,0]

#remove points with nan's
dcr = dc[~np.isnan(dc[:,1]),:]
jbr = jb[~np.isnan(jb[:,1]),:]
jmr = jm[~np.isnan(jm[:,1]),:]
uvr = uv[~np.isnan(uv[:,1]),:]

#function for linear interpolation
fdc = interpolate.interp1d(dcr[:,0],dcr[:,1],kind='linear')
fjb = interpolate.interp1d(jbr[:,0],jbr[:,1],kind='linear')
fjm = interpolate.interp1d(jmr[:,0],jmr[:,1],kind='linear')
fuv = interpolate.interp1d(uvr[:,0],uvr[:,1],kind='linear')

#replace nan's with interpolated values
dc[:,1] = fdc(tdc)
jb[:,1] = fjb(tjb)
jm[:,1] = fjm(tjm)
uv[:,1] = fuv(tuv)

#save to text files
np.savetxt('DCInterp.txt',dc)
np.savetxt('JBInterp.txt',jb)
np.savetxt('JMInterp.txt',jm)
np.savetxt('UVInterp.txt',uv)
```

### hour data

```python
#import data
h = np.genfromtxt('AllHourly.txt')

#save coordinates of stations
lon = h[:,0]-360. #lattitude
lat = h[:,1] #longitude
coords = np.array((lon,lat))
np.savetxt('HourlyCoords.txt',coords)

#isolate temperature data
temp = h[:,2:] #temperatures

#create time points in decimal days
time = np.arange(0,np.size(temp[0,:])*1./24.,1./24.)

N = np.size(lon)
for i in range(0,N):
    #remove points with nan's
    timer = time[~np.isnan(temp[i,:])]
    tempr = temp[i,~np.isnan(temp[i,:])]
    
    #linear interpolation
    f = interpolate.interp1d(timer,tempr,kind='linear',bounds_error=False,fill_value=np.mean(tempr[:2]))
    temp[i,:] = f(time) 

#save to text files
np.savetxt('HourlyInterp.txt',temp)
```

### Function to plot coastlines and station locations

```python
#arrows adds labels for minute stations
#geog adds geographic names
#legend adds a legend for the points

def plotMap(lonmin,lonmax,size,arrows=False,geog=False,legend=False):
    #set correct scale
    lonrange = lonmax-lonmin
    latrange = lonrange*np.cos(48.5*np.pi/180.)
    latmin = 48.3
    latmax = latmin+latrange
    
    #fig,ax = plt.subplots(1,1,figsize=(size,size))
    
    #plotcoastline data
    ax.plot(coastlon,coastlat,color='black',rasterized=True)
    
    #plot minute resolution stations and arrows
    for name,mlon,mlat,textx,texty in minstat:
        ax.plot(mlon,mlat,'.',marker='*',markersize=20.0,color='yellow',
                label='Minute Resolution Stations')
        if arrows==True:
            ax.annotate(
                name, xy=(mlon, mlat), xytext=(textx, texty),
                textcoords='offset points', fontsize=10,
                arrowprops=arrow_properties,)
    
    if geog==True:
        #annotate geography
        ax.text(-123.9,48.55,'Vancouver Island',fontsize='16')
        ax.text(-123.35,48.6,'Haro Strait',fontsize='13')
        ax.text(-123.97,48.34,'Juan De Fuca Strait',fontsize='13',rotation=-13)
        ax.text(-123.54,48.7,'Saanich Inlet',fontsize='13',rotation=-80)
    
    #plot hour resolution stations            
    ax.plot(hourlon,hourlat,'o',markersize=4.0,color='red',
        label='Hour Resolution Stations')
    
    if legend==True:
        #display minute resolution only once
        handles,labels = ax.get_legend_handles_labels()
        display = (0,4)
        ax.legend([handle for i, handle in enumerate(handles) if i in display],
                  [label for i, label in enumerate(labels) if i in display],
                  loc=2,fontsize='8',numpoints=1)
        
    #remove relative axis shift
    ax.get_xaxis().get_major_formatter().set_useOffset(False)
    #other plot formatting
    ax.set_xlabel('Longitude');ax.set_ylabel('Latitude');
    ax.set_xlim(lonmin,lonmax); ax.set_ylim(latmin,latmax);
    plt.tight_layout()
```

### Function to plot globally interpolated map

```python
def globalInterpMap(z,res,cbar=False,tmin=0,tmax=10):
    x = hourlon
    y = hourlat
    xi = np.linspace(np.amin(coastlon),np.amax(coastlon)+0.2,res)
    yi = np.linspace(np.amin(coastlat)-0.1,np.amax(coastlat),res)

    nx = len(xi)
    ny = len(yi)
    Zi = np.zeros((ny,nx))

    for i in range(nx):
        for j in range(ny):
            w = 1./((xi[i]-x)**2+(yi[j]-y)**2)
            w=w/np.sum(w)
            Zi[j,i]=np.sum(w*z)
    pcm = ax.pcolormesh(xi,yi,ma.masked_invalid(Zi))
    pcm.set_clim([tmin,tmax])
    
    if cbar==True:
        plt.colorbar(pcm,shrink=0.8)       
```

### Function to plot locally interpolated map (linear)

```python
def localInterpMap(z,res,cbar=False,tmin=0,tmax=10):
    x = hourlon
    y = hourlat
    xi = np.linspace(np.amin(coastlon),np.amax(coastlon)+0.2,res)
    yi = np.linspace(np.amin(coastlat)-0.1,np.amax(coastlat),res)
    Xx = np.vstack((x,y)).T
    Xi,Yi=np.meshgrid(xi,yi)
    Zi = interpolate.griddata(Xx,z,(Xi,Yi),method='linear')
    
    pcm = ax.pcolormesh(xi,yi,ma.masked_invalid(Zi))
    pcm.set_clim([tmin,tmax])
    
    if cbar==True:
        plt.colorbar(pcm,shrink=0.8)
```

### Function to tilt labels

```python
def tiltlab():
    labels = ax.get_xticklabels() 
    for label in labels: 
        label.set_rotation(30)
        label.set_ha('right')
```

## Low Pass the time series

The raw timeseries are very jagged due to higher frequency (daily patterns), to get a better picture of the overall yearly trends, the data was smoothed by filtering the data with an elliptical low pass filter, with a stopband of $3.5 \text{day}^{-1}$

###### Deep Cove Comparison Example

```python
day = 60.*24
df = 1./(3.5*day)

wp = 0.7*df
ws = 0.9*df

n,fn = signal.ellipord(wp,ws,0.5,20.)
b,a = signal.ellip(n,0.5,20.,fn)
dcf = signal.lfilter(b,a,dc[:,1])
dcf = signal.filtfilt(b,a,dc[:,1])

fig,ax = plt.subplots(1,1,figsize=(7.5,4))
ax.plot_date(dc[:,0],dcf,'-',color='blue',zorder=1,label='Low Pass')
ax.plot_date(dc[:,0],dc[:,1],'-',color='black',alpha=0.5,zorder=0,label='Raw Timeseries')
ax.legend(loc=0,fontsize='8')
ax.set_ylabel('Temperature [C]')
tiltlab()
fig.savefig('dclowpass.png',dpi=300)
```

###### Lowpass all four stations

```python
fig,ax = plt.subplots(1,1,figsize=(7.5,4))

for i in range(0,4):
    minutedata = files[i]
    temp = minutedata[:,1]
    t = minutedata[:,0]
    tempf = signal.lfilter(b,a,temp)
    tempf= signal.filtfilt(b,a,temp)
    ax.plot_date(t,tempf,'-',label=names[i])

ax.legend(loc=1,fontsize='8')
ax.set_ylabel('Temperatue [C]')
tiltlab()
fig.savefig('alllowpass.png',dpi=300)    
```

The smoothed timeseries clearly shows temperature increasing in the spring and summer months, and decreasing in the fall and winter months. This would be expected for locations in northern lattitudes.

John Muir elementary appears to be at a slightly lower temperature, particularily in the summer months. This can be investigated by looking at the probability distributions of the yearly temperature data, and by computing the univariate statistics of the raw timeseries.
